```python
#Se crea la clase con el motor de inferencia
class Saludos(KnowledgeEngine):
  @DefFacts() #conjunto de hechos
  def accion_inicial(self):
    yield Fact(accion = 'saluda')

  #Si no se tiene el nombre se pregunta y se guarda en la variable name
  @Rule(Fact(accion = 'saluda'), NOT(Fact(name=W())))
  def ask_name(self):
    self.declare(Fact(name=input('cual es tu nombre? ')))

  #Si no se tienen la ubicacion se pregunta y se guarda en la variable locacion
  @Rule(Fact(accion = 'saluda'), NOT(Fact(locacion=W())))
  def ask_location(self):
    self.declare(Fact(locacion=input('donde estas? ')))

  #Cuando se tiene nombre y ubicacion se saluda a la persona
  @Rule(Fact(accion = 'saluda'), Fact(name = 'name' <<W()),
        Fact(locacion = 'locacion' <<W()))
  def saludar(self, name, locacion):
    print('Hola %s! ¿como esta el clima en %s?'%(name, locacion))
```

```python
motor = Saludos()
motor.reset() #prepara el motor para ejecutar
motor.run()
```

```python
class RobotCrossStreet(KnowledgeEngine):
  @DefFacts()
  def accion_inicial(self):
    yield Fact(action = 'indicate')

  #Preguntas

  #Pregunta el color del semaforo (opcional)

  #@Rule(Fact(action = 'indicate'), NOT(Fact(light=W())))
  #def ask_light(self):
    #self.declare(Fact(light=input('Which color is the traffic light? ')))

  #Pregunta si vienen carros en caso de que el color del semaforo no sea rojo
  @Rule(Fact(action = 'indicate'), Fact(light = L('yellow')
  | L('blinking yellow') | L('green')), NOT(Fact(cars=W())))
  def ask_cars(self):
    self.declare(Fact(cars=input('Check both sides. Do you see cars? ')))

  #Pregunta si estan detenidos los carros
  @Rule(Fact(action = 'indicate'), Fact(light = L('yellow')
  | L('blinking yellow') | L('green')), Fact(cars = 'yes'),
  NOT(Fact(movement=W())))
  def ask_movement(self):
    self.declare(Fact(movement=input('Are those cars full stopped? ')))

  #Pregunta que tan cerca vienen los carros si la luz es verde
  @Rule(Fact(action = 'indicate'), Fact(light = 'green'), Fact(cars = 'yes'),
        Fact(movement = 'no'), NOT(Fact(near=W())))
  def ask_near(self):
    self.declare(Fact(near=input('do those moving cars are near you? ')))

  #Pregunta que tan rapido vienen los carros si la luz es verde
  @Rule(Fact(action = 'indicate'), Fact(light = 'green'), Fact(cars = 'yes'),
        Fact(movement = 'no'), Fact(near='no'),NOT(Fact(speed=W())))
  def ask_speed(self):
    self.declare(Fact(speed=input('do those moving cars are at high speed? ')))

  #Indicaciones
  #Indica que no se debe caminar si la luz es roja
  @Rule(Fact(action = 'indicate'), Fact(light = 'red'))
  def indicate(self):
    print('Light color is red. Dont walk')

  #Indica que se puede caminar si no hay carros
  @Rule(Fact(action = 'indicate'), Fact(light = 'light' <<W()),
        Fact(cars = 'no'))
  def indicateTwo(self, light):
    print('Light color is %s and no cars are coming. You can walk'%(light))

  #Indica que se puede caminar si los carros estan detenidos
  @Rule(Fact(action = 'indicate'), Fact(light = 'light' <<W()),
        Fact(cars = 'yes'), Fact(movement = 'yes'))
  def indicateThree(self, light):
    print('Light color is %s and the cars are full stopped.'
    ' You can walk'%(light))

  #Indica que se puede caminar si la luz es verde, los carros estan lejos y
  #no van a alta velocidad
  @Rule(Fact(action = 'indicate'), Fact(light = 'green'), Fact(cars = 'yes'),
        Fact(movement = 'no'), Fact(near = 'no'), Fact(speed = 'no'))
  def indicateFour(self):
    print('Light color is green, the cars are far and not at hight speed.'
          ' You can walk but still be cautious')

  #Indica que se debe esperar si los carros se estan moviendo muy cerca si la
  #luz es verde
  @Rule(Fact(action = 'indicate'), Fact(light = 'green'), Fact(cars = 'yes'),
        Fact(movement = 'no'), Fact(near = 'yes'))
  def indicateFive(self):
    print('Light color is green, the cars are close to you.'
          ' Better wait unitl they full stop or until they pass the crossing')

  #Indica que se debe esperar si la luz es amarilla y los carros siguen en
  #movimiento
  @Rule(Fact(action = 'indicate'), Fact(light = L('yellow')
  | L('blinking yellow')), Fact(cars = 'yes'), Fact(movement = 'no'))
  def indicateSix(self):
    print('Light color is yellow and the cars are moving.'
    ' Better wait to the next green light')

  #Indica que no se puede caminar si los carros van a alta velocidad
  @Rule(Fact(action = 'indicate'), Fact(light = 'green'), Fact(cars = 'yes'),
        Fact(movement = 'no'), Fact(near = 'near' <<W()), Fact(speed = 'yes'))
  def indicateSeven(self):
    print('Light color is green and the cars are at high speed. Dont walk.'
    ' Wailt until they are full stopped or until they pass the crossing')
```

```python
engine = RobotCrossStreet()
engine.reset() #prepara el motor para ejecutar
engine.declare(Fact(light=choice(['green', 'yellow',
                                   'blinking yellow', 'red'])))
```

```python
engine.run()
```
